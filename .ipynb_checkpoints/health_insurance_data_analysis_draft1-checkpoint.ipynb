## 3 Medicare

### 3.1 Load Data set

State/County Level containing cost/expenditure [Link](https://www.cms.gov/Research-Statistics-Data-and-Systems/Statistics-Trends-and-Reports/Medicare-Geographic-Variation/GV_PUF) <<State County All Table 2018.xlsx>>

HRR Level containing cost/expenditure [Link](https://www.cms.gov/Research-Statistics-Data-and-Systems/Statistics-Trends-and-Reports/Medicare-Geographic-Variation/GV_PUF)

```python
%%time # use time measurement to check the load time of data source file = 3250 * 248 table
medicare_beneficiaries = pd.read_excel("data/State County All Table 2018.xlsx", sheet_name="State_county 2018", header=1, index_col=None)
medicare_beneficiaries.head()
```

```python
# DataFrame Column Labels Formatter
medicare_beneficiaries.rename(columns = lambda x:x.strip().upper(), inplace=True)
medicare_beneficiaries.rename(columns = {"STATE":"STATE ABBREVIATION"}, inplace=True)
```

```python
medicare_beneficiaries.head()
```

### 3.2 MEDICARE STATE LEVEL STATISTICS
We can use "STATE TOTAL" to filter state-level statistics from medicare data

```python
# Remove the noise data from dataset "UNKONWN" rows
medicare_beneficiaries = medicare_beneficiaries[medicare_beneficiaries.COUNTY != "UNKNOWN"]
# PICK up state level statistics
medicare_beneficiaries_by_state = medicare_beneficiaries[medicare_beneficiaries["COUNTY"] == "STATE TOTAL"]
medicare_beneficiaries_by_state = medicare_beneficiaries_by_state[
    medicare_beneficiaries_by_state.columns[[0, 3, 4, 5]]]
```

```python
df_state_population = pd.merge(df_state_result, medicare_beneficiaries_by_state, on="STATE ABBREVIATION")
df_state_population.info()
```

```python
df_state_population["BENEFICIARIES WITH PART A AND PART B"] = df_state_population["BENEFICIARIES WITH PART A AND PART B"].astype(int)
df_state_population["FFS BENEFICIARIES"] = df_state_population["FFS BENEFICIARIES"].astype(int)
df_state_population["MA BENEFICIARIES"] = df_state_population["MA BENEFICIARIES"].astype(int)
```

## 4 Conclusion
We have got a result dataframe of # of state-level population, # of medicaid enrollment, # of medicare beneficiaries. By utilizing these statistic data, we compute % of "STATE_MEDICAID_PRCENT" = medicaid enrollment / state-level population, % of "STATE_MEDICARE_PRCENT" = # of medicare beneficiaries / state-level population. 

Suppose % of private provider customers = 1 - % of "STATE_MEDICAID_PRCENT" - % of "STATE_MEDICARE_PRCENT

[Potential Provider list](https://www.cms.gov/Research-Statistics-Data-and-Systems/Downloadable-Public-Use-Files/Provider-of-Services)

```python
# Choose "TOTAL MEDICAID AND CHIP ENROLLMENT" as medicaid total
df_state_population["STATE_MEDICAID_PRCENT"] = df_state_population["TOTAL MEDICAID AND CHIP ENROLLMENT"] / df_state_population["TOT_POP"]
df_state_population["STATE_MEDICARE_PRCENT"] = df_state_population["BENEFICIARIES WITH PART A AND PART B"] / df_state_population["TOT_POP"]
df_state_population["STATE_PRIVATE_PRCENT"] = 1 - (df_state_population["STATE_MEDICAID_PRCENT"] + df_state_population["STATE_MEDICARE_PRCENT"])
df_state_pop = df_state_population[["STATE ABBREVIATION", "STNAME", "YEAR", "STATE_MEDICAID_PRCENT", 
                                   "STATE_MEDICARE_PRCENT", "STATE_PRIVATE_PRCENT"]]
df_state_pop.info()
```

## 5 Price/Fee Schedule Data

1. ***CPT/HCPCT Code Set***
> 1. [CPT Code](https://www.cms.gov/Research-Statistics-Data-and-Systems/Downloadable-Public-Use-Files/Part-B-National-Summary-Data-File/Overview)
The data sets are summarized by meaningful ***Healthcare Common Procedure Coding/Current Procedural Terminology, (HCPC/CPT)***, code ranges. Brief descriptions for the code ranges and modifiers are provided in the readme file. The data set name contains the year followed by a five character sequence that is the HCPC/CPT code. This HCPC/CPT code corresponds to the first HCPC/CPT in the selected code range of disciplines.
> 2. [HCPCT Codeset](https://www.cms.gov/Medicare/Coding/HCPCSReleaseCodeSets/Alpha-Numeric-HCPCS)These files contain the Level II alphanumeric HCPCS procedure and modifier codes, their long and short descriptions, and applicable Medicare administrative, coverage and pricing data

2. ***Physician Fee Schedule***
> * **Medicare PFS** <br> It is located in CMS physician fee schedule [overview](https://www.cms.gov/apps/physician-fee-schedule/search/search-criteria.aspx) tool. Here is the [guideline/specification](https://www.cms.gov/apps/physician-fee-schedule/help/How_to_MPFS_Booklet_ICN901344.pdf) for the searchable tool.<img src="images/HowToSearchPFS.png">
> * **Medicaid PFS** <br> It can be retrieved from each state's healthcare official website. Below are three samples from **Indiana**, **Connecticut**, and **Illinois**.
>> 1. [Indiana medicaid phsician fee schedule](http://provider.indianamedicaid.com/ihcp/Publications/MaxFee/fee_home.asp)
>> 2. [Connecticut medicaid phsician fee schedule](https://www.ctdssmap.com/CTPortal/Provider/ProviderFeeScheduleDownload/tabid/54/Default.aspx)
>> 3. [Illinois medicaid phsician fee schedule](https://www.illinois.gov/hfs/MedicalProviders/MedicaidReimbursement/Pages/Practitioner.aspx)

```python
#Indiana medicaid physician fee schedule
df_medicaid_OPFS_IN = pd.read_excel("data/Indiana/Outpatient Fee Schedule 2020-04.xlsx", sheet_name="Tab 3 - Fee Schedule", 
                                       header=16, index_col=None, usecols=None)
df_medicaid_OPFS_IN.rename(columns = lambda x:x.strip().upper(), inplace=True)
df_medicaid_OPFS_IN = df_medicaid_OPFS_IN[df_medicaid_OPFS_IN["FEE SCHED AMT"].notnull()]
df_medicaid_OPFS_IN["PRICE EFFECTIVE"] = df_medicaid_OPFS_IN["PRICE EFFECTIVE"].apply(pd.to_datetime)
df_medicaid_OPFS_IN["PROC CODE"] = df_medicaid_OPFS_IN["PROC CODE"].apply(str)
df_medicaid_OPFS_IN = df_medicaid_OPFS_IN[df_medicaid_OPFS_IN["PRICE EFFECTIVE"] < pd.to_datetime("1/1/2019")]
```

```python
df_medicaid_OPFS_IN.info()
```

```python
df_medicaid_OPFS_IN.head()
```

```python
#Medicare Searchable CPT Pricing info for certain CPT
#
df_medicare_PriceInfo_IN = pd.read_csv("data/Indiana/PFSExport_IN_2018.csv", header=0, 
                                       index_col=None, usecols=[0, 2, 5, 6, 7, 8])
df_medicare_PriceInfo_IN.rename(columns = lambda x:x.strip().upper(), inplace=True)
df_medicare_PriceInfo_IN.info()
```

```python
#Handle $12,234.56 char formatted cash amount to float type
for col in ("NON-FACILITY PRICE", "FACILITY PRICE"):
    df_medicare_PriceInfo_IN[col] = df_medicare_PriceInfo_IN[col].str.replace(r',', '')
df_medicare_PriceInfo_IN["NON-FACILITY PRICE"] = (df_medicare_PriceInfo_IN["NON-FACILITY PRICE"]
                                                  .str.replace(r'[^-+\d.]', '').astype(float))
df_medicare_PriceInfo_IN["FACILITY PRICE"] = (df_medicare_PriceInfo_IN["FACILITY PRICE"]
                                                  .str.replace(r'[^-+\d.]', '').astype(float))
```

```python
df_medicare_PriceInfo_IN.rename(columns={"HCPCS CODE":"PROC CODE"}, inplace=True)
```

```python
df_medicare_PriceInfo_IN.head()
```

### Combination of both Medicaid and medicare price info

```python
df_CMS_PriceInfo_IN = pd.merge(df_medicaid_OPFS_IN, df_medicare_PriceInfo_IN, on="PROC CODE")
```

```python
df_CMS_PriceInfo_IN.info()
```

### Compute the anchor price

```python
medicaidPrcent = df_state_pop.iloc[14, 3]
print(type(medicaidPrcent))
medicarePrcent = df_state_pop.iloc[14, 4]
print(type(medicarePrcent))
privatePrcent = df_state_pop.iloc[14, 5]
print(type(privatePrcent))
df_CMS_PriceInfo_IN["ANCHOR PRICE"] = df_CMS_PriceInfo_IN["FEE SCHED AMT"]*medicaidPrcent + \
    df_CMS_PriceInfo_IN["NON-FACILITY PRICE"]*medicarePrcent + \
    df_CMS_PriceInfo_IN["NON-FACILITY PRICE"]*privatePrcent
```

### Formatted Output Dataframe

```python
df_PRICE_IN = df_CMS_PriceInfo_IN[["PROC CODE", "DESCRIPTION", "FEE SCHED AMT", "NON-FACILITY PRICE", "ANCHOR PRICE"]]
df_PRICE_IN.rename(columns={"FEE SCHED AMT":"MEDICAID_PFS_AMT", "NON-FACILITY PRICE":"MEDICARE_PFS_AMT"}, inplace=True)
df_PRICE_IN.head(50)
```
